```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# OJO: este tiene un espacio en el nombre, lo corregimos as√≠
df_tienda1 = pd.read_csv('tienda_1 .csv')
df_tienda2 = pd.read_csv('tienda_2.csv')
df_tienda3 = pd.read_csv('tienda_3.csv')
df_tienda4 = pd.read_csv('tienda_4.csv')

# Mostramos las primeras filas de cada uno para confirmar que se cargaron bien
print("üìä Tienda 1:")
display(df_tienda1.head())

print("üìä Tienda 2:")
display(df_tienda2.head())

print("üìä Tienda 3:")
display(df_tienda3.head())

print("üìä Tienda 4:")
display(df_tienda4.head())
```

```python
ingresos = {}  # Creamos un diccionario vac√≠o para guardar los ingresos por tienda

for i in range(1, 5):  # Bucle que recorre desde 1 hasta 4 (porque tienes tiendas 1 a 4)
    df = globals()[f'df_tienda{i}']  # Usa el nombre de la variable df_tienda1, df_tienda2, etc.
    ingresos[f'Tienda {i}'] = df['Precio'].sum()  # Calcula el ingreso total y lo guarda en el diccionario

# Mostrar los resultados formateados
for tienda, ingreso in ingresos.items():
    print(f'üõí Ingreso total {tienda}: ${ingreso:,.2f}')
```

```python
# Encuentra la tienda con mayor ingreso
tienda_top = max(ingresos, key=ingresos.get)
ingreso_top = ingresos[tienda_top]

# Muestra el resultado
print(f'üèÜ La tienda con mayor ingreso es {tienda_top} con un total de ${ingreso_top:,.2f}')
```

```python
# Encontrar la tienda con el menor ingreso
tienda_menor = min(ingresos, key=ingresos.get)
ingreso_menor = ingresos[tienda_menor]

# Mostrar resultado
print(f' La tienda con menor ingreso es {tienda_menor} con un total de ${ingreso_menor:,.2f}')
```

```python
# Calcular el promedio de ingresos
promedio_ingreso = sum(ingresos.values()) / len(ingresos)

# Mostrar el resultado
print(f'üìä Promedio de ingreso entre las 4 tiendas: ${promedio_ingreso:,.2f}')
```

```python
# Funci√≥n para mostrar productos vendidos por categor√≠a sin gr√°ficos ni metadata
def productos_por_categoria(df, nombre_tienda):
    conteo = df['Categor√≠a del Producto'].value_counts()
    print(f"\nüìä Categor√≠as m√°s populares en {nombre_tienda}:\n")
    for categoria, cantidad in conteo.items():
        print(f"{categoria}: {cantidad}")


# Ejecutar para cada tienda
productos_por_categoria(df_tienda1, 'Tienda 1')
productos_por_categoria(df_tienda2, 'Tienda 2')
productos_por_categoria(df_tienda3, 'Tienda 3')
productos_por_categoria(df_tienda4, 'Tienda 4')
```

```python
# Funci√≥n para identificar los productos m√°s y menos vendidos por tienda
def productos_destacados(df, nombre_tienda):
    conteo = df['Producto'].value_counts()

    producto_top = conteo.idxmax()      # Producto m√°s vendido
    cantidad_top = conteo.max()

    producto_min = conteo.idxmin()      # Producto menos vendido
    cantidad_min = conteo.min()

    print(f"\nüõí {nombre_tienda}")
    print(f"üîù Producto m√°s vendido: {producto_top} ({cantidad_top} ventas)")
    print(f"üîª Producto menos vendido: {producto_min} ({cantidad_min} venta{'s' if cantidad_min > 1 else ''})")

# Ejecutar para cada tienda
productos_destacados(df_tienda1, 'Tienda 1')
productos_destacados(df_tienda2, 'Tienda 2')
productos_destacados(df_tienda3, 'Tienda 3')
productos_destacados(df_tienda4, 'Tienda 4')
```

```python
# ‚úÖ Preparar listas de datos
tiendas = list(ingresos.keys())
valores = list(ingresos.values())

# ‚úÖ Crear gr√°fico de barras
plt.figure(figsize=(8, 5))
barras = plt.bar(tiendas, valores, color='skyblue')

# ‚úÖ Agregar etiquetas encima de cada barra
for i, barra in enumerate(barras):
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2,
             altura + max(valores) * 0.01,
             f"${valores[i]:,.0f}",
             ha='center', va='bottom', fontsize=10)

# ‚úÖ T√≠tulos y ejes
plt.title('Ingreso total por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso total ($)')
plt.tight_layout()
plt.show()
```

```python
# Funci√≥n para graficar las categor√≠as m√°s vendidas de una tienda
def graficar_categorias(df, nombre_tienda):
    conteo = df['Categor√≠a del Producto'].value_counts()

    plt.figure(figsize=(10, 5))
    conteo.plot(kind='barh', color='skyblue')
    plt.title(f' {nombre_tienda} ‚Äì Productos vendidos por categor√≠a')
    plt.xlabel('Cantidad de productos vendidos')
    plt.ylabel('Categor√≠a')
    plt.tight_layout()
    plt.show()

# Ejecutar para cada tienda
graficar_categorias(df_tienda1, 'Tienda 1')
graficar_categorias(df_tienda2, 'Tienda 2')
graficar_categorias(df_tienda3, 'Tienda 3')
graficar_categorias(df_tienda4, 'Tienda 4')
```

```python
# Funci√≥n para graficar los productos m√°s vendidos de una tienda
def graficar_productos_top(df, nombre_tienda, top_n=10):
    conteo = df['Producto'].value_counts().head(top_n)

    plt.figure(figsize=(10, 5))
    conteo.plot(kind='barh', color='skyblue')
    plt.title(f'{nombre_tienda} ‚Äì Top {top_n} productos m√°s vendidos')
    plt.xlabel('Cantidad de ventas')
    plt.ylabel('Producto')
    plt.gca().invert_yaxis()  # Mostrar el producto m√°s vendido arriba
    plt.tight_layout()
    plt.show()

# Ejecutar para cada tienda
graficar_productos_top(df_tienda1, 'Tienda 1')
graficar_productos_top(df_tienda2, 'Tienda 2')
graficar_productos_top(df_tienda3, 'Tienda 3')
graficar_productos_top(df_tienda4, 'Tienda 4')
```
